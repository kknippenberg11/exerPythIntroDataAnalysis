```python
data = pd.read_csv("/content/exerPythIntroDataAnalysis/countries.csv")
print(data)
```

```python
data
```

```python
data.country == "United States"
```

```python
data[data.country == "United States"]
```

```python
#Compare population growth in the US vs China.
us = data[data.country == "United States"]
china = data[data.country == "China"]
us
```

```python
plt.plot(us.year,us.population / 10**6)
plt.plot(us.year,china.population / 10**6)
plt.xlabel("year")
plt.ylabel("Millions")
plt.legend(["US", "China"]);
```

```python
us.population
```

```python
us_change = us.population/us.population.iloc[0]
us_change
```

```python
china_change = china.population/china.population.iloc[0]
china_change
```

```python
plt.plot(us.year,us_change)
plt.plot(us.year,china_change)
plt.xlabel("Year")
plt.ylabel("% Change")
plt.legend(["US", "China"]);
```
